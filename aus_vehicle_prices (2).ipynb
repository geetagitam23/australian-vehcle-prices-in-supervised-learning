This dataset contains the latest information on car prices in Australia for the year 2023. It covers various brands, models, types, and features of cars sold in the Australian market. It provides useful insights into the trends and factors influencing the car prices in Australia. The dataset has over 16,000 records of car listings from various online platforms in Australia.

```python
import pandas as pd

def fill_outliers_with_iqr(df, column):
    # Convert column to numeric
    df[column] = pd.to_numeric(df[column], errors='coerce')

    # Calculate quartiles and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with NaN
    df[column] = df[column].apply(lambda x: x if (x >= lower_bound and x <= upper_bound) else pd.NA)

    # Fill NaN values with median
    median = df[column].median()
    df[column].fillna(median, inplace=True)

    return df

# Assuming 'data' is your DataFrame
data = fill_outliers_with_iqr(df, column='Price')
```

# Model & Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score

# Function for Regression Report
def regression_report_train_test(y_train, y_train_pred, y_test, y_test_pred):
    print('''
            ------------------------------------------
            |     REGRESSION REPORT FOR TRAIN DATA   |
            ------------------------------------------
            ''')
    print("Mean Squared Error (Train):", mean_squared_error(y_train, y_train_pred))
    print("R-squared (Train):", r2_score(y_train, y_train_pred))

    print('''
            ------------------------------------------
            |      REGRESSION REPORT FOR TEST DATA   |
            ------------------------------------------
            ''')
    print("Mean Squared Error (Test):", mean_squared_error(y_test, y_test_pred))
    print("R-squared (Test):", r2_score(y_test, y_test_pred))

# Defining error metrics
scores_regression = pd.DataFrame(columns=['Model', 'Train_MSE', 'Train_R2', 'Test_MSE', 'Test_R2'])

def get_regression_metrics(train_actual, train_predicted, test_actual, test_predicted, model_description, dataframe):
    train_mse = mean_squared_error(train_actual, train_predicted)
    train_r2 = r2_score(train_actual, train_predicted)
    test_mse = mean_squared_error(test_actual, test_predicted)
    test_r2 = r2_score(test_actual, test_predicted)

    dataframe = dataframe.append(pd.Series([model_description, train_mse, train_r2, test_mse, test_r2],
                                           index=scores_regression.columns), ignore_index=True)
    return dataframe
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression

# Linear Regression
reg_lr = LinearRegression()
reg_lr.fit(X_train, y_train)

# Predictions
y_pred_train_lr = reg_lr.predict(X_train)
y_pred_test_lr = reg_lr.predict(X_test)

# Train Metrics
train_mse = mean_squared_error(y_train, y_pred_train_lr)
train_r2 = r2_score(y_train, y_pred_train_lr)

print("Train Mean Squared Error (MSE):", train_mse)
print("Train R-squared (R2):", train_r2)

# Test Metrics
test_mse = mean_squared_error(y_test, y_pred_test_lr)
test_r2 = r2_score(y_test, y_pred_test_lr)

print("Test Mean Squared Error (MSE):", test_mse)
print("Test R-squared (R2):", test_r2)

# Regression Report
regression_report_train_test(y_train, y_pred_train_lr, y_test, y_pred_test_lr)

# Metrics
scores_regression = get_regression_metrics(y_train, y_pred_train_lr, y_test, y_pred_test_lr, "Linear Regression", scores_regression)

# Confusion Matrix (not applicable for regression)
# If needed, add code for plotting confusion matrix for classification
```
